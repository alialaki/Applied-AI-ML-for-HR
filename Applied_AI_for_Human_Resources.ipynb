#**Next, we provide data for machine learning.**
#1- We first covert the dataset into a NumPy array of type float.
#This is the preferred input formant for Keras
#2- we split the feature and target variables into X & Y.

```python
#Convert to Dataframe to numpy array
np_attrition = attrition_data.to_numpy().astype(float)

#Create X_train with the first 7 attributes
X_train = np_attrition[:,1:7]

# #Create Y_train with attrition atribute
Y_train = np_attrition[:,7]

#Convert Y_train to one-hot encoding
#Since the attrition is bolean datatype, it has two unique values
Y_train = tf.keras.utils.to_categorical(Y_train,2)
print(f"X_train shape: {X_train.shape}")
print(f"Y_train shape: {Y_train.shape}")
```

#02.05. Build Attrition model with Keras

```python
from os import name
from tensorflow import keras
from tensorflow.keras import optimizers
from tensorflow.keras.regularizers import l2

#Setup hyperprameter for deep learning
EPOCHS = 100
BATCH_SIZE = 100
#we can see the result if verbose set to 1. if I set to 0 it'll hide the result
VERBOSE = 1
#two calsses
NB_CLASSES = 2
N_HIDDEN = 128
#This means 20% of the data will be used for validation while building the model
VALIDATION_SPLIT = 0.2

#Create a Keras model
model = tf.keras.models.Sequential()

#Add first hidden Dense Layer
model.add(keras.layers.Dense(N_HIDDEN,
                             input_shape=(6,),
                             name='Dense-Layer-1',
                             activation='relu'))
model.add(keras.layers.Dense(N_HIDDEN,
                             name='Dense-layer-2',
                             activation='relu'))

#Add a final layer with softmax
model.add(keras.layers.Dense(NB_CLASSES,
                             name='Final',
                             activation='softmax'))
#Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

#Fit parameters
model.fit(X_train,
          Y_train,
          batch_size=BATCH_SIZE,
          epochs=EPOCHS,
          verbose=VERBOSE,
          validation_split=VALIDATION_SPLIT)
```

#**O2.06. Predict Attrition with Keras**

```python
attrition_data.columns
```

```python
TotalMonthsOfExp=40
TotalOrgsWorked=4
MonthsInOrg=20
LastPayIncrementBand=5
AverageFeedback=4
LastPromotionYears=4

# Create a NumPy array from the input list
input_data = np.array([[TotalMonthsOfExp, TotalOrgsWorked, MonthsInOrg,
       LastPayIncrementBand, AverageFeedback, LastPromotionYears]])

# Use model.predict to get the predicted probabilities
prediction = model.predict(input_data)

# Get the predicted class by finding the index of the maximum probability
predicted_class = np.argmax(prediction, axis=1)

print("Will employee leave ? (0: No Attrition, 1: Attrition)?", predicted_class[0])
```
